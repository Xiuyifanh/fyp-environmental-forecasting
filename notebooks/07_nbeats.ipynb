```python
# ============================================================
# Notebook 07 — N-BEATS Modelling
# Baseline V1: Interpretable stack, default configuration
# Reference: Oreshkin et al. (2020) ICLR
# Implementation: Nixtla neuralforecast library
# Author: Nihaar Raut | 001309432
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from neuralforecast import NeuralForecast
from neuralforecast.models import NBEATS
from sklearn.metrics import mean_absolute_error, mean_squared_error
import warnings
warnings.filterwarnings('ignore')

PROCESSED = Path('../data/processed')
FIGURES = Path('../outputs/figures')

# Load all four series
ndvi_amazon_train = pd.read_csv(PROCESSED / 'ndvi_amazon_train.csv', parse_dates=['date'])
ndvi_amazon_test = pd.read_csv(PROCESSED / 'ndvi_amazon_test.csv', parse_dates=['date'])

ndvi_sahel_train = pd.read_csv(PROCESSED / 'ndvi_sahel_train.csv', parse_dates=['date'])
ndvi_sahel_test = pd.read_csv(PROCESSED / 'ndvi_sahel_test.csv', parse_dates=['date'])

sst_atlantic_train = pd.read_csv(PROCESSED / 'sst_atlantic_train.csv', parse_dates=['date'])
sst_atlantic_test = pd.read_csv(PROCESSED / 'sst_atlantic_test.csv', parse_dates=['date'])

sst_indian_train = pd.read_csv(PROCESSED / 'sst_indian_train.csv', parse_dates=['date'])
sst_indian_test = pd.read_csv(PROCESSED / 'sst_indian_test.csv', parse_dates=['date'])

print("All series loaded successfully")
```

```python
# ============================================================
# N-BEATS V1 — Baseline, interpretable stack
# Input size: 24 months (2 years lookback)
# Forecast horizon: 48 months (test period)
# Interpretable stack: trend + seasonality basis expansion
# neuralforecast requires specific dataframe format
# ============================================================

def prepare_nbeats_df(train, value_col, series_id):
    df = train[['date', value_col]].copy()
    df.columns = ['ds', 'y']
    df['unique_id'] = series_id
    df['ds'] = pd.to_datetime(df['ds'])
    return df

def run_nbeats(train, test, value_col, series_name, series_id):
    # Prepare data
    train_df = prepare_nbeats_df(train, value_col, series_id)
    
    # Define model
    model = NBEATS(
    h=48,
    input_size=24,
    stack_types=['trend', 'seasonality'],
    n_blocks=[3, 3],
    mlp_units=[[256, 256], [256, 256]],
    n_harmonics=2,
    n_polynomials=2,
    max_steps=500,
    learning_rate=1e-3,
    random_seed=42
    )
    
    nf = NeuralForecast(models=[model], freq='MS')
    nf.fit(train_df)
    
    # Forecast
    forecast_df = nf.predict()
    forecast = forecast_df['NBEATS'].values
    actual = test[value_col].values[:len(forecast)]
    
    # Metrics
    mae = mean_absolute_error(actual, forecast)
    rmse = np.sqrt(mean_squared_error(actual, forecast))
    
    # Plot
    fig, ax = plt.subplots(figsize=(14, 4))
    ax.plot(train['date'], train[value_col],
            color='#2c3e50', linewidth=1, label='Train')
    ax.plot(test['date'], actual,
            color='#27ae60', linewidth=1.5, label='Actual')
    
    forecast_dates = pd.date_range(
        start=test['date'].iloc[0], periods=len(forecast), freq='MS')
    ax.plot(forecast_dates, forecast,
            color='#c0392b', linewidth=1.5, linestyle='--', label='N-BEATS Forecast')
    ax.axvline(x=test['date'].iloc[0], color='grey',
               linestyle=':', linewidth=1, label='Train/Test Split')
    ax.set_title(f'N-BEATS V1 — {series_name} | MAE: {mae:.4f} | RMSE: {rmse:.4f}',
                 fontweight='bold')
    ax.set_xlabel('Date')
    ax.legend()
    ax.grid(True, alpha=0.3)
    plt.tight_layout()
    
    filename = f"fig_nbeats_v1_{series_name.lower().replace(' ', '_')}.png"
    plt.savefig(FIGURES / filename, dpi=150, bbox_inches='tight')
    plt.show()
    
    return {'series': series_name, 'MAE': round(mae, 4), 'RMSE': round(rmse, 4)}

# Run V1 on all four series
results = []

print("Training N-BEATS — this will take several minutes...")
print()

results.append(run_nbeats(ndvi_amazon_train, ndvi_amazon_test,
                           'ndvi_mean', 'NDVI Amazon', 'ndvi_amazon'))

results.append(run_nbeats(ndvi_sahel_train, ndvi_sahel_test,
                           'ndvi_mean', 'NDVI Sahel', 'ndvi_sahel'))

results.append(run_nbeats(sst_atlantic_train, sst_atlantic_test,
                           'sst_mean', 'SST North Atlantic', 'sst_atlantic'))

results.append(run_nbeats(sst_indian_train, sst_indian_test,
                           'sst_mean', 'SST Indian Ocean', 'sst_indian'))

# Results table
results_df = pd.DataFrame(results)
print("\n=== N-BEATS V1 Baseline Results ===")
print(results_df.to_string(index=False))
```

```python
# ============================================================
# N-BEATS V2 — Tuned architecture with interpretable stack
# Improvements over V1 (default NeuralForecast settings):
#   - Increased input_size from 24 to 36 (3x seasonal period)
#   - Increased n_blocks and mlp_units for richer representation
#   - Interpretable stack: trend + seasonality decomposition
#   - max_steps increased from 100 to 500 for better convergence
#   - random_seed fixed for reproducibility
# Reference: Oreshkin et al. (2020) ICLR, arxiv:1905.10437
# ============================================================

from neuralforecast import NeuralForecast
from neuralforecast.models import NBEATS
from neuralforecast.losses.pytorch import MAE as NBEATS_MAE
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

def prepare_nbeats_df(train, value_col, freq='MS'):
    """Convert to NeuralForecast long format."""
    df = train[['date', value_col]].copy()
    df.columns = ['ds', 'y']
    df.insert(0, 'unique_id', 'series_1')
    return df

def run_nbeats_v2(train, test, value_col, series_name):
    print(f"\nFitting N-BEATS V2 for {series_name}...")
    
    train_df = prepare_nbeats_df(train, value_col)
    
    # N-BEATS V2 — interpretable stack, tuned architecture
    models = [NBEATS(
        h=len(test),                    # forecast horizon = test length
        input_size=36,                  # 3 years lookback (3x m=12)
        stack_types=['trend', 'seasonality'],  # interpretable decomposition
        n_blocks=[3, 3],               # 3 blocks per stack
        mlp_units=[[512, 512],
                   [512, 512]],        # wider MLP layers
        n_harmonics=2,                 # Fourier terms for seasonality
        n_polynomials=3,               # polynomial degree for trend
        max_steps=500,                 # more training steps vs V1
        learning_rate=1e-3,
        loss=NBEATS_MAE(),
        random_seed=42,
        enable_progress_bar=False
    )]
    
    nf = NeuralForecast(models=models, freq='MS')
    nf.fit(train_df)
    
    # Forecast
    future_df = pd.DataFrame({
        'unique_id': 'series_1',
        'ds': test['date'].values
    })
    
    forecast_df = nf.predict()
    forecast = forecast_df['NBEATS'].values
    actual = test[value_col].values
    
    # Metrics
    mae = mean_absolute_error(actual, forecast)
    rmse = np.sqrt(mean_squared_error(actual, forecast))
    
    # Plot
    fig, ax = plt.subplots(figsize=(14, 4))
    ax.plot(train['date'], train[value_col],
            color='#2c3e50', linewidth=1, label='Train')
    ax.plot(test['date'], actual,
            color='#27ae60', linewidth=1.5, label='Actual')
    ax.plot(test['date'], forecast,
            color='#9b59b6', linewidth=1.5, linestyle='--',
            label='N-BEATS V2 Forecast')
    ax.axvline(x=test['date'].iloc[0], color='grey',
               linestyle=':', linewidth=1, label='Train/Test Split')
    ax.set_title(f'N-BEATS V2 — {series_name} | MAE: {mae:.4f} | RMSE: {rmse:.4f}',
                 fontweight='bold')
    ax.set_xlabel('Date')
    ax.legend()
    ax.grid(True, alpha=0.3)
    plt.tight_layout()
    
    filename = f"fig_nbeats_v2_{series_name.lower().replace(' ', '_')}.png"
    plt.savefig(FIGURES / filename, dpi=150, bbox_inches='tight')
    plt.show()
    
    return {
        'series': series_name,
        'input_size': 36,
        'max_steps': 500,
        'MAE': round(mae, 4),
        'RMSE': round(rmse, 4)
    }

# Run V2 on all four series
results_nbeats_v2 = []

results_nbeats_v2.append(run_nbeats_v2(
    ndvi_amazon_train, ndvi_amazon_test,
    'ndvi_mean', 'NDVI Amazon'))

results_nbeats_v2.append(run_nbeats_v2(
    ndvi_sahel_train, ndvi_sahel_test,
    'ndvi_mean', 'NDVI Sahel'))

results_nbeats_v2.append(run_nbeats_v2(
    sst_atlantic_train, sst_atlantic_test,
    'sst_mean', 'SST North Atlantic'))

results_nbeats_v2.append(run_nbeats_v2(
    sst_indian_train, sst_indian_test,
    'sst_mean', 'SST Indian Ocean'))

# Results table
results_nbeats_v2_df = pd.DataFrame(results_nbeats_v2)
print("\n=== N-BEATS V2 Results ===")
print(results_nbeats_v2_df.to_string(index=False))
```
